```python
import numpy as np
import pandas as pd

# npy 파일 경로
file_path = './data/S01_test_X.npy'

# npy 파일을 numpy 배열로 읽기
data = np.load(file_path)
# 배열의 모양 변경: (288, 1, 22, 1125) -> (288, 22*1125)
# 이는 첫 번째 차원을 유지하고 나머지 차원을 합칩니다.
print(data.shape)
reshaped_data = data.reshape(data.shape[0], -1)
# numpy 배열을 DataFrame으로 변환
df = pd.DataFrame(reshaped_data)

print(df)
```
